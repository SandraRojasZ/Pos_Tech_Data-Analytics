```python
df_diabetes = pd.read_csv("https://github.com/SandraRojasZ/Machine-Learning-and-Time-Series/blob/main/Base_de_Dados/diabetes.csv?raw=true", sep=",")
```

```python
import pandas as pd
```

```python
! pip install scikit-learn
```

```python
df_diabetes.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_diabetes.drop(["Class variable"], axis=1)
```

```python
x.head(3)
```

```python
y = df_diabetes["Class variable"]
```

```python
y.head(3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
len(x_train)
```

```python
df_diabetes.shape
```

```python
len(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinar
knn.fit(x_train, y_train)
```

```python
accuracy = knn.score(x_test, y_test)
```
